```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import random

from google.colab import files
from datetime import datetime
from IPython.display import display

def unqFileName(base):
  now = datetime.now()
  # dd/mm/YY H:M:S
  dt_string = now.strftime("%d%m%Y_%H%M%S")
  return base + "_" + dt_string + ".csv"

def downloadOrDisplay(file, df, head=False, both=False):  
  download = ""
  while both==False and download != "Y" and download != "N":
    download = input("download? [y/n] --> ").upper()
  if both or download == "Y":
    fNm = unqFileName(file)
    df.to_csv(fNm, encoding='utf-8') 
    files.download(fNm)
  if both or download != "Y":
    if head:
      display(df.head())
    else:
      display(df)
```

```python
#don't work: linkedin, thesaurus
def wordHippo(word):
  url = "https://www.wordhippo.com/what-is/another-word-for/"+word+".html"
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  defs = [val.text for val in soup.find_all('div', class_ = "tabdesc")]
  wordLs = [[word.text for word in words.find_all('div', class_ = "wb")] for words in soup.find_all('div', class_ = "relatedwords")] #includes 2 empty lists at the end for some reason

  thesaurus = {}
  for i in range(min(len(defs), len(wordLs))):
    thesaurus[defs[i]] = wordLs[i]
  return thesaurus

def resumeCompanion():
  #url = "https://www.indeed.com/career-advice/resumes-cover-letters/resume-adjectives" #didn't work
  url = "https://resumecompanion.com/resume-help/resume-adjectives/"
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  words = [td.text for td in soup.find_all('td')]
  return words

thes = {'word':[], 'i':[]}
defs = {'word':[], 'i':[], 'def':[]}

wordn = 50
print('-'*wordn)
for wordi, word in enumerate(random.sample(resumeCompanion(), wordn)):
  print('-',end='')
  thesaurus = wordHippo(word)
  
  for i, key in enumerate(thesaurus.keys()): #thesaurus = definition -> [word, word, word]
    thes['word'].append(word)
    thes['i'].append(i)
    
    defs['word'].append(word)
    defs['i'].append(i)
    
    for syni, syn in enumerate(thesaurus[key]):
      if thes.get('syn'+str(syni), 0) == 0:
        thes['syn'+str(syni)] = [pd.NA for i in range(len(thes['word'])-1)]
      thes['syn'+str(syni)].append( thesaurus[key][syni] )

    defs['def'].append(key)

thes_df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in thes.items() ]))
defs_df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in defs.items() ]))

downloadOrDisplay("thesaurus", thes_df, both=True)
downloadOrDisplay("definition", defs_df, both=True)
#data uploaded here https://drive.google.com/drive/folders/18dEvVuQrfJRZP2pSZLhCIHLQMAJomKVr?usp=sharing
```
