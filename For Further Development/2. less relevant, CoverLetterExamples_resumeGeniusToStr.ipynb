```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

#don't work: linkedin, thesaurus
profLetters = "https://resumegenius.com/cover-letter-examples"
soup = BeautifulSoup(requests.get(profLetters).text, 'html.parser')
letterDivs = [div.find('div', class_="entry-content") for div in soup.find_all('div', class_="accordion-section accordion-section-shortcode")]

for letterDiv in letterDivs:
  print("\n".join([p.text for p in letterDiv.find_all('p')]))
  print()
```
